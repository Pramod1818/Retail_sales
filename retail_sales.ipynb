```python
import pandas as pd
df = pd.read_csv('orders.csv', na_values=(['Not Available','unknown'])) #these  2 treat as null
df.head(10)
```

```python
df['Ship Mode'].unique()
```

```python
# df.rename(columns={'Order Id':'order_id','City':'city'}) but this is no good for renaming 15 cols
df.columns= df.columns.str.lower()
```

```python
df.columns= df.columns.str.replace(' ','_')
df.columns
```

```python
df['total_discount']=df['list_price']*df['discount_percent']*.01*df['quantity']
```

```python
df.head()
```

```python

df['total_sale']=(df['list_price']*df['quantity'])-df['total_discount']
```

```python
df['total_profit']=  (df['total_sale']-df['cost_price']*df['quantity'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
df.drop(columns={'list_price','cost_price','discount_percent'},inplace=True)
```
